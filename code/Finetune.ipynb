```python
from ultralytics import YOLO
```

# Train with VisDrone dataset

```python
# Load a model
model = YOLO("yolo11s.pt")

# Train the model
results = model.train(
    data="/content/drive/MyDrive/ZALO_AI/VisDrone_val_Yolo/data.yaml",
    epochs=50,
    imgsz=960,
    batch=16,
    resume=True,

    scale=0.5,
    degrees=15,
    mixup=0.2,
    project="/content/drive/MyDrive/ZALO_AI/VisDrone_val_Yolo/yolo11s"
)
```
